```python
from collections import deque

def find_paths(grid):
    h, w = len(grid), len(grid[0])
    starts = {'@': None}
    walls = set()

    # Parse the input map
    for y in range(h):
        for x in range(w):
            if grid[y][x].islower() or grid[y][x] == '@':
                starts[grid[y][x]] = (x, y)
            elif grid[y][x] == '#':
                walls.add((x, y))

    # Helper function to find paths
    def bfs(start, end):
        queue = deque([(starts[start], '', 0)]) # (position, capitals, distance)
        visited = set()
        while queue:
            (x, y), capitals, dist = queue.popleft()
            if (x, y) == starts[end]:
                return (dist, capitals[:-1])
            for dx, dy in [(0, 1), (1, 0), (0, -1), (-1, 0)]:
                nx, ny = x + dx, y + dy
                if 0 <= nx < w and 0 <= ny < h and (nx, ny) not in walls and (nx, ny) not in visited:
                    visited.add((nx, ny))
                    new_capitals = capitals + grid[ny][nx] if grid[ny][nx].isalpha() else capitals
                    queue.append(((nx, ny), new_capitals, dist + 1))
        return None

    # Building the output dictionary
    paths = {}
    for start in starts:
        for end in starts:
            if start != end and end != '@':
                path = bfs(start, end)
                if path:
                    paths[(start, end)] = path

    return paths

paths = find_paths(map)
```

```python
def shortest_path(map_dict, visited, current_node):
    # Base case: If all lowercase letters are visited
    if all(chr(ord('a') + i) in visited for i in range(26)):
        return 0

    # Function caching to optimize for repeated calls
    @lru_cache(maxsize=None)
    def dfs(node, visited):
        min_distance = float('inf')

        # Iterating through all possible next steps
        for (start, end), (distance, required_nodes) in map_dict.items():
            if start == node and end not in visited:
                # Check if all required uppercase letters have been visited
                if all(rn.lower() in visited for rn in required_nodes):
                    new_visited = visited | {end}
                    min_distance = min(min_distance, distance + dfs(end, frozenset(new_visited)))

        return min_distance if min_distance != float('inf') else 0

    return dfs(current_node, frozenset(visited))

# Testing the function with the updated map and starting from '@'
shortest_path(paths, set(), '@')
```

```python
from collections import deque

def find_paths(grid):
    h, w = len(grid), len(grid[0])
    starts = {}
    walls = set()
    at_counter = 1  # Counter for '@'

    # Parse the input map
    for y in range(h):
        for x in range(w):
            if grid[y][x] == '@':
                # Assign a unique identifier for each '@'
                starts[f'@{at_counter}'] = (x, y)
                at_counter += 1
            elif grid[y][x].islower():
                starts[grid[y][x]] = (x, y)
            elif grid[y][x] == '#':
                walls.add((x, y))

    # Helper function to find paths
    def bfs(start, end):
        queue = deque([(starts[start], '', 0)])  # (position, capitals, distance)
        visited = set()
        while queue:
            (x, y), capitals, dist = queue.popleft()
            if (x, y) == starts[end]:
                return (dist, capitals[:-1])
            for dx, dy in [(0, 1), (1, 0), (0, -1), (-1, 0)]:
                nx, ny = x + dx, y + dy
                if 0 <= nx < w and 0 <= ny < h and (nx, ny) not in walls and (nx, ny) not in visited:
                    visited.add((nx, ny))
                    new_capitals = capitals + grid[ny][nx] if grid[ny][nx].isalpha() else capitals
                    queue.append(((nx, ny), new_capitals, dist + 1))
        return None

    # Building the output dictionary
    paths = {}
    for start in starts:
        for end in starts:
            if start != end and not end.startswith('@'):
                path = bfs(start, end)
                if path:
                    paths[(start, end)] = path

    return paths

paths = find_paths(map)
```

```python
from functools import lru_cache

def shortest_path_multiple_starts(map_dict, visited, current_nodes):
    # Base case: If all lowercase letters are visited
    if all(chr(ord('a') + i) in visited for i in range(26)):
        return 0

    # Function caching to optimize for repeated calls
    @lru_cache(maxsize=None)
    def dfs(visited, current_nodes):
        min_distance = float('inf')

        # Iterating through each current node and possible next steps
        for current_node in current_nodes:
            for (start, end), (distance, required_nodes) in map_dict.items():
                if start == current_node and end not in visited:
                    # Check if all required uppercase letters have been visited
                    if all(rn.lower() in visited for rn in required_nodes):
                        new_visited = visited | {end}
                        new_current_nodes = set(current_nodes) - {current_node} | {end}
                        min_distance = min(min_distance, distance + dfs(frozenset(new_visited), frozenset(new_current_nodes)))

        return min_distance if min_distance != float('inf') else 0

    return dfs(frozenset(visited), frozenset(current_nodes))

shortest_path_multiple_starts(paths, set(), {'@1', '@2', '@3', '@4'})
```
