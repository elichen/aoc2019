```python
def get_val(index, mode, base, arr):
    if mode == 0:
      return arr.get(arr.get(index,0),0)
    elif mode == 1:
      return arr.get(index,0)
    elif mode == 2:
      return arr.get(arr.get(index,0)+base,0)

def run(input, a, i=0, base=0):
    while i < len(a):
        m3, m2, m1, op = a[i] // 10000, a[i] // 1000 % 10, a[i] // 100 % 10, a[i] % 100
        a1 = get_val(i + 1, m1, base, a)
        a2 = get_val(i + 2, m2, base, a)

        if op == 1:
            addr = a[i+3]+base if m3 == 2 else a[i+3]
            a[addr] = a1 + a2
            i += 4
        elif op == 2:
            addr = a[i+3]+base if m3 == 2 else a[i+3]
            a[addr] = a1 * a2
            i += 4
        elif op == 3:
            if input == None: return input, None, i, base
            addr = a[i+1]+base if m1 == 2 else a[i+1]
            a[addr] = input
            input = None
            i += 2
        elif op == 4:
            return input, a1, i+2, base
        elif op == 5:
            i = a2 if a1 != 0 else i + 3
        elif op == 6:
            i = a2 if a1 == 0 else i + 3
        elif op == 7:
            addr = a[i+3]+base if m3 == 2 else a[i+3]
            a[addr] = 1 if a1 < a2 else 0
            i += 4
        elif op == 8:
            addr = a[i+3]+base if m3 == 2 else a[i+3]
            a[addr] = 1 if a1 == a2 else 0
            i += 4
        elif op == 9:
            base += a1
            i += 2
        elif op == 99:
            return input, None, -99, -99
```

```python
# Find the shape of the array
rows, cols = m.shape

# Initialize an empty list to store the nodes that meet the condition
target_nodes = []

# Iterate through each element in the array
for i in range(rows):
    for j in range(cols):
        # Skip if the current node is not '#'
        if m[i, j] != '#':
            continue

        # Initialize a counter for '#' neighbors
        count = 0

        # Check UDLR directions
        directions = [(i-1, j), (i+1, j), (i, j-1), (i, j+1)]
        for x, y in directions:
            if 0 <= x < rows and 0 <= y < cols:
                if m[x, y] == '#':
                    count += 1

        # Check if the node is surrounded by 3 or more '#' neighbors
        if count >= 3:
            target_nodes.append((i, j))
```

```python
import numpy as np

# Convert the string representation of the map into a numpy array
map_array = m

# Directions
directions = {'up': (-1, 0), 'right': (0, 1), 'down': (1, 0), 'left': (0, -1)}
left_turn = {'up': 'left', 'right': 'up', 'down': 'right', 'left': 'down'}
right_turn = {'up': 'right', 'right': 'down', 'down': 'left', 'left': 'up'}

# Starting position and direction
start_pos = tuple((map_array == '^').nonzero())
facing = 'up'
current_pos = start_pos

# Function to check the next move
def next_move(pos, direction):
    return pos[0] + direction[0], pos[1] + direction[1]

# Function to get the length of the path ahead
def path_length(pos, direction):
    steps = 0
    while True:
        pos = next_move(pos, direction)
        if 0 <= pos[0] < map_array.shape[0] and 0 <= pos[1] < map_array.shape[1] and map_array[pos] == '#':
            steps += 1
        else:
            break
    return steps

# Traversal
commands = []
while True:
    dir_vector = directions[facing]
    if path_length(current_pos, dir_vector) > 0:  # Move forward if path ahead
        steps = path_length(current_pos, dir_vector)
        commands.append(str(steps))
        current_pos = next_move(current_pos, (steps * dir_vector[0], steps * dir_vector[1]))
    else:  # At a dead end or no path forward, try turning
        left_dir = left_turn[facing]
        right_dir = right_turn[facing]
        if path_length(current_pos, directions[left_dir]) > 0:  # Can turn left
            commands.append('L')
            facing = left_dir
        elif path_length(current_pos, directions[right_dir]) > 0:  # Can turn right
            commands.append('R')
            facing = right_dir
        else:  # No paths left or right, end traversal
            break

# Print out the list of commands
print(','.join(commands))
```
